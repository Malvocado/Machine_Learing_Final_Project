Import Library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('BMW_sales_data_after_processing.csv')
```

Filter year feature to only in range 2022-2024, because in there are too many record in the previous dataset

```python
df = df[(df['Year'] >= 2022) & (df['Year'] <= 2024)].reset_index(drop=True)
display(df.head())
```

Check df's shape after filtered

```python
print(f"Shape of the filtered DataFrame: {df.shape}")
```

Check Unique value

```python
print(f"Unique years in the filtered DataFrame: {df['Year'].unique()}")
```

## Dataset Analysis & Transformation
We will analyze the 'Year' column distribution and then calculate the age of the records relative to the year 2026.

### Convert Year to Relative Age
Transforming the year format from a specific year (e.g., 2022) to the number of years relative to 2024 (Highest year in the dataset).

```python
HIGHEST_YEAR = 2024
df['Years_Relative'] = HIGHEST_YEAR - df['Year']

print(f"Sample of transformation (Relative to {CURRENT_YEAR}):")
display(df[['Year', 'Years_Relative']].head())
```

```python
print(f"Filtered Feature: ",df['Years_Relative'].head())
print(f"\nWhole dataset: ", df.head())
```

###Droping region column by filter dataset to only asia

```python
# Filter the dataset to only include 'Asia'
df = df[df['Region'] == 'Asia']

# Drop the 'Region' column
df = df.drop(columns=['Region'])
df = df.reset_index(drop=True)

# Display the results
print(f"Shape after filtering for Asia: {df.shape}")
display(df.head())
```

### Final Result

```python
print(f"Dataset Shape:",df.shape)
print(f"All columns:",df.columns.to_list())
print(df.head())
```
